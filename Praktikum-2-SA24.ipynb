### Contoh 2 â€” Mengecek Angka Genap atau Ganjil

```python
def cek_genap_ganjil(bilangan):
    """Memeriksa apakah bilangan genap atau ganjil dan mencetak hasilnya."""
    if bilangan % 2 == 0:
        print(f"Angka {bilangan} adalah **Genap**.")
    else:
        print(f"Angka {bilangan} adalah **Ganjil**.")

# Contoh Penggunaan:
cek_genap_ganjil(10)
cek_genap_ganjil(15)
```

**Latihan 2:** Buatlah fungsi baru bernama `cek_positif_negatif` yang menerima satu input angka dan mencetak pesan `Angka [X] adalah Positif` atau `Angka [X] adalah Negatif atau Nol`.

### Contoh 3 â€” Function Menghitung Luas Persegi Panjang

```python

def hitung_luas_persegipanjang(panjang, lebar):
    """Menghitung luas (P * L)."""
    luas = panjang * lebar
    return luas

# Contoh Penggunaan:
p = 5
l = 8
hasil_luas = hitung_luas_persegipanjang(p, l)
print(f"Luas persegi panjang (5x8) adalah {hasil_luas}")
```

Buatlah fungsi baru bernama `hitung_keliling_persegipanjang` yang menerima input `panjang` dan `lebar`, dan mengembalikan (return) nilai kelilingnya ($\text{Keliling} = 2 \times (\text{P} + \text{L})$).

### Contoh 4 â€” Function Perhitungan Diskon Harga

```python
def hitung_harga_diskon(harga_awal, persentase_diskon):
    """Menghitung harga akhir setelah diskon diberikan."""
    # Ubah persentase diskon ke bentuk desimal
    diskon = persentase_diskon / 100
    jumlah_diskon = harga_awal * diskon
    return jumlah_diskon

# Contoh Penggunaan:
harga = 500000 # 500 ribu
diskon = 20    # 20%
jumlah_diskon = hitung_harga_diskon(harga, diskon)
print(f"Harga awal Rp 500.000 dengan diskon 20% adalah Rp {jumlah_diskon:,.0f}")
```

Buatlah fungsi baru bernama `hitung_jumlah_bayar` yang menerima input `harga_awal` dan `persentase_diskon`, dan mengembalikan (return) jumlah diskon  (misalnya, diskon anda Rp 100.000) serta harga akhirnya (misalnya, yang harus dibayar adalah Rp 400.000 ).

### ðŸ”¹ Contoh 7 â€” Function dengan input dan return

```python
def hitung_diskon(premi, persen):
    return premi - (premi * persen / 100)

hasil = hitung_diskon(1000000, 10)
print(f"Premi setelah diskon: Rp {hasil:,}")
```

### ðŸ”¹ Contoh 8

```python
def hitung_bunga(principal, rate, tahun):
    bunga = principal * (rate/100) * tahun
    return bunga

hasil = hitung_bunga(1000000, 5, 2)
print(f"Total bunga selama 2 tahun: Rp {hasil:,}")
```

### ðŸ”¹ Contoh 9

```python
def hitung_nilai_aktual(nilai_masa_depan, bunga, waktu):
    return nilai_masa_depan / ((1 + bunga) ** waktu)

nilai = hitung_nilai_aktual(1000000, 0.05, 3)
print(f"Nilai sekarang: Rp {nilai:,.2f}")
```

### ðŸ”¹ Contoh 10

```python
def nilai_harapan(probabilitas, hasil):
    total = 0
    for i in range(len(probabilitas)):
        total += probabilitas[i] * hasil[i]
    return total

p = [0.1, 0.3, 0.6]
x = [100000, 200000, 500000]
print(f"Nilai harapan: Rp {nilai_harapan(p, x):,.2f}")
```
