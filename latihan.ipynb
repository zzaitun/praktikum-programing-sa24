$f(x) = x^2 - x + \sqrt{x}$

```python
def fun(x):
    y = x**2 - x + x**(1/2)
    print(f"nilai dari f({x}) adalah {y}")

x = float(input("Masukkan nilai x= "))

fun(x)
```

$f(x,y) = \frac{x+y}{x^2+y}$

```python
def fun2(x,y):
    fxy = (x+y)/(x**2+y)
    print(f"Nilai dari f({x},{y}) adalah {fxy}")


fun2(x=2.1,y=1.1)
```
